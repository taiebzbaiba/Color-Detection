```python
# Installing dependancies

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Imporing image for testing

img=cv2.imread('C:\\Users\\taieb\\Desktop\\color ident\\full_color.jpg')
```

```python
#Plotting the testing image with BGR format

plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
#Converting BGR to RGB

grid_RGB =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# Plotting our RGB image

plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
# Initialising grid_Hsv variable

grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
#Indexing upper and lower values for hsv 
#Identifying the yellow colour in this case


lower = np.array([25,150,50])
upper = np.array([30,255,255])
```

```python
#Creating Image Mask with selected HSV value limits 

mask = cv2.inRange(grid_HSV , lower , upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(grid_RGB, grid_RGB ,mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```
